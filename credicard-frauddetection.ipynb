```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme(style='dark')
from tensorflow import keras
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# reading csv file
creditCard_dataset = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
creditCard_dataset.head(10)

# checking the frequency of each class
creditCard_dataset['Class'].value_counts()

# Dividing the dataset into dependent and independent variables
X, Y = creditCard_dataset.drop('Class', axis=1), creditCard_dataset['Class']

# SMOTE - to oversample minority class
sm = SMOTE(random_state=42)
X_res, Y_res = sm.fit_resample(X, Y)
# Creating a new dataframe after resampling
resampled_credit_dataset = pd.DataFrame(X_res, columns=X_res.columns)
resampled_credit_dataset['Class'] = Y_res
resampled_credit_dataset.head(10)

resampled_credit_dataset['Class'].value_counts()

# Selecting the best features according to highest correlation value
resampled_credit_dataset.corrwith(resampled_credit_dataset['Class'])

sorted(abs(resampled_credit_dataset.corrwith(resampled_credit_dataset['Class'])), reverse=True)
# V14, V4, V11, V10, V16, V9, V3
X = np.array(resampled_credit_dataset.drop('Class', axis=1))
Y = np.array(resampled_credit_dataset['Class'])
print(X)

# training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=420)

# Adding the layers to the model
ann = keras.Sequential([
keras.layers.Input(shape=(30,)),
keras.layers.Flatten(),
keras.layers.Dense(4, activation='relu'),
keras.layers.Dense(2, activation='relu'),
keras.layers.Dense(1, activation='sigmoid')])

# Compiling the model
ann.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])

# Fitting the training data in the model
ann.fit(X_train, Y_train, epochs=15)

# Predicting classes
Y_prob = ann.predict(X_test)
Y_pred = Y_prob.argmax(axis=1)
print(Y_pred)

# Displaying the classification report
print(classification_report(Y_test, Y_pred))

```
