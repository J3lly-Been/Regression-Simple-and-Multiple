```python
# -*- coding: utf-8 -*-
"""PythonWorkshop_CaseStudy1.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1Vv1sUy7c0U4LEh2VCbhMrSVfhQj586lk

# DATA RETRIEVAL
"""

# IMPORTING MODULES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# OPENING THE CSV FILES
hdf = pd.read_csv("../input/banks-analysis/HDFCBANK.csv")
ici = pd.read_csv("../input/banks-analysis/ICICIBANK.csv")
kot = pd.read_csv("../input/banks-analysis/KOTAKBANK.csv")

# HDFC DATAFRAME
hdf.head(-1)

# ICICI DATAFRAME
ici.head(-1)

# KOTAK DATAFRAME
kot.head(-1)

"""ALL DATA IS TILL 29th APRIL 2021

# DATA FORMAT ANALYSIS
"""

# STATISTICAL DESCRIPTION HDFC BANK
hdf.describe()

# DATAFRAME INFO HDFC BANK
hdf.info()

# STATISTICAL DESCRIPTION ICICI BANK
ici.describe()

# DATAFRAME INFO ICICI BANK
ici.info()

# STATISTICAL DESCRIPTION KOTAK BANK
kot.describe()

# DATAFRAME INFO KOTAK BANK
kot.info()

"""# DATA WRANGLING

HIGH/LOW
"""

# ALL TIME HIGH/LOW HDFC BANK vs PREVIOUS 2 YEARS
hdf[['Date','High','Low']].plot(x='Date',y=['High','Low'])
hdf[['Date','High','Low']].tail(730).plot(x='Date',y=['High','Low'])

# ALL TIME HIGH/LOW ICICI BANK vs PREVIOUS 2 YEARS
ici[['Date','High','Low']].plot(x='Date',y=['High','Low'])
ici[['Date','High','Low']].tail(730).plot(x='Date',y=['High','Low'])

# ALL TIME HIGH/LOW KOTAK BANK vs PREVIOUS 2 YEARS
kot[['Date','High','Low']].plot(x='Date',y=['High','Low'])
kot[['Date','High','Low']].tail(730).plot(x='Date',y=['High','Low'])

# FORMATTING FOR 2 YEAR HIGHS
dates = list(hdf['Date'].tail(730))
hdf_high = list(hdf['High'].tail(730))
ici_high = list(ici['High'].tail(730))
kot_high = list(kot['High'].tail(730))

# FORMATTING FOR 2 YEAR VOLUME
hdf_vol = list(hdf['Volume'].tail(730))
ici_vol = list(ici['Volume'].tail(730))
kot_vol = list(kot['Volume'].tail(730))

# VOLUMES COMPARED (FOR CONTEXT)
plt.plot(dates,hdf_vol,label='HDFC')
plt.plot(dates,ici_vol,label='ICICI')
plt.plot(dates,kot_vol,label='KOTAK')
plt.legend()

# HIGHS COMPARED
plt.plot(dates,hdf_high,label='HDFC')
plt.plot(dates,ici_high,label='ICICI')
plt.plot(dates,kot_high,label='KOTAK')
plt.legend()

"""STOCK LIQUIDITY"""

# ALL TIME TURNOVER/TRADES HDFC vs PREVIOUS 2 YEARS
print("ALL TIME")
print(hdf[['Turnover','Trades']].describe())
print("")
print("LAST 2 YEARS")
print(hdf[['Turnover','Trades']].tail(730).describe())

# ALL TIME TURNOVER/TRADES ICICI vs PREVIOUS 2 YEARS
print("ALL TIME")
print(ici[['Turnover','Trades']].describe())
print("")
print("LAST 2 YEARS")
print(ici[['Turnover','Trades']].tail(730).describe())

# ALL TIME TURNOVER/TRADES KOTAK vs PREVIOUS 2 YEARS
print("ALL TIME")
print(kot[['Turnover','Trades']].describe())
print("")
print("LAST 2 YEARS")
print(kot[['Turnover','Trades']].tail(730).describe())

# FORMATTING FOR 2 YEAR LQUIDITY
hdf_turn = list(hdf['Trades'].tail(730))
ici_turn = list(ici['Trades'].tail(730))
kot_turn = list(kot['Trades'].tail(730))

# TRADES COMPARED
plt.plot(dates,hdf_turn,label='HDFC')
plt.plot(dates,ici_turn,label='ICICI')
plt.plot(dates,kot_turn,label='KOTAK')
plt.legend()

"""TRADES MADE vs TRADES DELIVERED"""

# ALL TIME VOL/DEL_VOL HDFC vs PREVIOUS 2 YEARS
print("ALL TIME")
print(hdf[['Volume','Deliverable Volume','%Deliverble']].describe())
print("")
print("LAST 2 YEARS")
print(hdf[['Volume','Deliverable Volume','%Deliverble']].tail(730).describe())

# ALL TIME VOL/DEL_VOL ICICI vs PREVIOUS 2 YEARS
print("ALL TIME")
print(ici[['Volume','Deliverable Volume','%Deliverble']].describe())
print("")
print("LAST 2 YEARS")
print(ici[['Volume','Deliverable Volume','%Deliverble']].tail(730).describe())

# ALL TIME VOL/DEL_VOL KOTAK vs PREVIOUS 2 YEARS
print("ALL TIME")
print(kot[['Volume','Deliverable Volume','%Deliverble']].describe())
print("")
print("LAST 2 YEARS")
print(kot[['Volume','Deliverable Volume','%Deliverble']].tail(730).describe())

# ALL TIME HIGH/LOW HDFC BANK vs PREVIOUS 2 YEARS
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].plot(x='Date',y=['Volume','Deliverable Volume'])
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].tail(730).plot(x='Date',y=['Volume','Deliverable Volume'])

# ALL TIME HIGH/LOW HDFC BANK vs PREVIOUS 2 YEARS
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].plot(x='Date',y=['Volume','Deliverable Volume'])
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].tail(730).plot(x='Date',y=['Volume','Deliverable Volume'])

# ALL TIME HIGH/LOW HDFC BANK vs PREVIOUS 2 YEARS
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].plot(x='Date',y=['Volume','Deliverable Volume'])
hdf[['Date','Volume','Deliverable Volume','%Deliverble']].tail(730).plot(x='Date',y=['Volume','Deliverable Volume'])

# FORMATTING FOR 2 YEAR %DELIVERABLE
date_60 = list(hdf['Date'].tail(60))
hdf_pdev = list(hdf['%Deliverble'].tail(60))
ici_pdev = list(ici['%Deliverble'].tail(60))
kot_pdev = list(kot['%Deliverble'].tail(60))

# %DELIVERABLE COMPARED
plt.plot(date_60,hdf_pdev,label='HDFC')
plt.plot(date_60,ici_pdev,label='ICICI')
plt.plot(date_60,kot_pdev,label='KOTAK')
plt.legend()
```
